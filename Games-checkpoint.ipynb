### Вступление

**Название проекта**: Успешность игры.   
    
**Описание проекта**: Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».
    
**Цель проекта**: Выявить определяющие успешность игры закономерности.  

**Описание данных**:

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Есть 4 рейтинга в которых осуществляется продажа почти всех игр, это E, E10+, M и T. Игры с другими рейтингами почти не продаются. Рейтинг E является самым привлекательным для пользователей.

```python
#Проверяем гипотезу, что средние пользовательские рейтинги Xbox One и PC одинаковые.
rating_x_pc = data.pivot_table(index = 'platform', values = 'user_score', aggfunc = ['mean','median'])
rating_x_pc.columns = ['mean', 'median']
display(rating_x_pc.query('platform == ["PC","XOne"]'))
```

Нулевая гипотеза: средний пользовательский рейтинг платформ Xbox One и PC одинаковый. Альтернативная гипотеза: средний пользовательский рейтинг платформ Xbox One и PC разный

```python
data_p = data.loc[data['user_score'] > 0.1]
```

```python
# Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC одинаковы
p_pc = data_p[data_p['platform'] == 'PC']['user_score']
p_x = data_p[data_p['platform'] == 'XOne']['user_score']
result = st.ttest_ind(p_pc, p_x, equal_var = False)
alpha = 0.05

print(result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нулевая гипотеза: средний пользовательский рейтинг жанров Action и Sports одинаковый. Альтернативная гипотеза: средний пользовательский рейтинг жанров Action и Sports разный.

```python
# Проверка гипотезы: средние пользовательские рейтинги жанров Action и Sports разные.
g_action = data_p[data_p['genre'] == 'Action']['user_score']
g_sports = data_p[data_p['genre'] == 'Sports']['user_score']
results = st.ttest_ind(g_action, g_sports, equal_var = False)
alpha = 0.05

print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

В топ продажах:PS2, X360, Wii, PS3, DS.Так же было выяснено, что платформы живут около 7 лет. Потенциально прибыльные платформы: XOne, PSV, PS4. В основном предпочитают консоли PS и XB. При покупки игры внимание больше обращают на оценки критиков чем пользователей, но это не значит, что если оценки низкие игры не наберет по продажам. Самые прибыльные Shooter игры, из топ стабильно прибыльных жанров по продажам так же можно отметить Platform, Sports, Racing, Misc. Было выяснено, что жители Северной Америки, Европы и других стран использует примерно одни и те же платформы в отличие от граждан Японии у которых вкусы значительно различаются. Так же в Японии любят жанр Role – Playing который намного меньше ценится в других регионах. Вкусы пользователей Северной Америки и Европы по жанрам схожи. Игры с рейтингом E продаются лучше всего, меньше продаются игры с рейтингами E10+, M и T, игры с другими рейтингами почти не продаются. Так же были проверены две гипотезы которые были отвергнуты.
